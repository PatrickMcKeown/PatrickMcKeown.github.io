Private construction spending deteriorated significantly in the United States during the Great Recession, which according to the Bureau of Labor Statistics, lasted from December 2007 to June 2009.   Private construction spending began recovery in June 2011.

```python
fig,ax = plt.subplots()
x = df_Construction['Month-Year']
private = ax.plot(df_Construction['Month-Year'], df_Construction["Private Construction"], alpha = 0.75, c = 'b')
ax.legend()
#public = ax.plot(df_Construction['Month-Year'], df_Construction["Public Construction"], alpha = 0.5, c = 'k')
#ax.legend(['Private Construction', 'Public Construction'])

fig.suptitle('Private Construction Spending Impact with the Great Recession', fontsize = '16', fontname = 'DejaVu Sans',)
ax.xaxis.set_label_text('Month/Year', fontsize = '14', fontname = 'DejaVu Sans')
ax.yaxis.set_label_text('$ Construction Spending (in millions)', fontsize = '14', fontname = 'DejaVu Sans')
ax.set_xticks(x[::6])

#ax.axvline(x=114,linewidth=1,color='k',label='Private Construction Spending Begins to Recover',linestyle='--')
ax.axvspan(72,90,ymin=0,ymax = 90000,facecolor='k',alpha=0.1)
ax.axvspan(114, 146, ymin=0, ymax = 90000, facecolor='g', alpha=0.1)
ax.text(81,75000,'The Great Recession',fontsize=12,verticalalignment='center',horizontalalignment='center',fontname='DejaVu Sans')
ax.text(130,62000,'Private Construction Spending\nBegins to Recover', fontsize=12, verticalalignment='center', horizontalalignment='center', fontname = 'DejaVu Sans')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.set_size_inches(20,14)
fig.savefig('Private Construction Spending and the Great Recession.jpg', dpi=600)
```
